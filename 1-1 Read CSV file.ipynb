```python
#使用yfinance套件下載股票資料)
import yfinance as yf
import pandas as pd
start_date ="2020-01-01"
end_date="2022-03-05"
ticker="2603.TW"                                         #股票名
data=yf.download(ticker,start_date,end_date)
data.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
data["Adj Close"].plot(figsize=(10,7))
plt.title("Evergreen",fontsize=16)
plt.ylabel('Price',fontsize=14)
plt.xlabel('Year',fontsize=14)
plt.grid(which="major",color="k",linestyle="-",linewidth=0.5)
plt.show()
```

```python
#定義函數與for迴圈指令
   #運用定義函數與for迴圈指令,製作產品邀請信
def product_msg(customers):
    str1="123"                     #字串(str)
    str2="456"
    str3="789"
    str4="1235"
    for customer in customers:
         msg=str1+customer+"\n"+str2+"\n"+str3+"\n"+str4
    print(msg,'\n')
member=["a","b","c"]
product_msg(member)
```

```python
#流程控制(迴圈)決策      #景氣對策信號分數
score=int(input("請輸入景氣對策信號分數:"))
if score>=45 or score<9:
    print("錯誤")                     #如左列 if...條件.....然後輸入":"換行,寫print....
elif score>=37:                       # 一定要換行
    print("紅燈")
elif score>=37:                       #選取起來按Tab可縮排
    print("紅燈")
elif score>=31:
    print("黃紅燈")
elif score>=22:
    print("綠燈")
elif score>=16:
    print("黃藍燈")
elif score>=9:
    print("藍")
```
